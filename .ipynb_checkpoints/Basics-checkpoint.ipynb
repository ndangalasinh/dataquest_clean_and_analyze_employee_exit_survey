# Data analysis of the Exit interview at DETE and TAFE

This is the analysis of the data from the exit interviews of employees of DETE and TAFE

There are two data sets from each institute, the first one is dete_survey.csv
and the second one is tafe_survey.csv

```python
#importing pandas and NumPy libraries to be used in the project.
import pandas as pd
import numpy as np
```

```python
#Reading the CSV files into pandas and assigning them into respective variables
dete_survey = pd.read_csv('dete_survey.csv')
tafe_survey = pd.read_csv('tafe_survey.csv')
```

```python
dete_survey = pd.read_csv('dete_survey.csv',na_values= 'Not Stated')
```

```python
dete_survey_updated = dete_survey.drop(dete_survey.columns[28:49],axis =1)
```

```python
tafe_survey_updated = tafe_survey.drop(tafe_survey.columns[17:66], axis = 1)
```

### In the above code lines we have dropped the columns that are not very useful in our analysis.

```python
dete_survey_updated.columns = dete_survey_updated.columns.str.replace(' ','_').str.strip().str.lower()
dete_survey_updated.columns
```

```python
tafe_survey_updated = tafe_survey_updated.rename(columns={'Record ID': 'id','CESSATION YEAR': 'cease_date','Reason for ceasing employment': 'separationtype','Gender. What is your Gender?': 'gender','CurrentAge. Current Age': 'age','Employment Type. Employment Type': 'employment_status','Classification. Classification': 'position','LengthofServiceOverall. Overall Length of Service at Institute (in years)': 'institute_service'})
```

### Now we are going to make a copy of the main dataframe selecting only those rows that will help us answer the questions presented to us.

```python
dete_resignations = dete_survey_updated[(dete_survey_updated['separationtype']=='Resignation-Other reasons')|(dete_survey_updated['separationtype']=='Resignation-Other employer')|(dete_survey_updated['separationtype']=='Resignation-Move overseas/interstate')].copy()
```

```python
tafe_resignations = tafe_survey_updated[tafe_survey_updated['separationtype']=='Resignation'].copy()
```
