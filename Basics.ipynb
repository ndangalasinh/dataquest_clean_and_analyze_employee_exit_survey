```python
#This is the formula we will be using to remove all the '-'
def update_vals(element):
    if element == '-':
        return False
    elif pd.isnull(element):
        return np.nan
    else:
        return True
```

```python
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction','Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(axis = 1, skipna = False)
```

#### Combine the dataframes. Assign the result to combined

```python
combined = pd.concat([dete_resignations_up,tafe_resignations_up],ignore_index= True)
```

```python
#check for bad data
combined.notnull().sum().sort_values()
```

```python
combined_updated = combined.dropna(axis=1,thresh= 500)
```

```python
combined_updated
```

```python
combined_updated['institute_service'].value_counts(dropna = False)
```

```python
#extracting the information about the institute_service
combined_updated['institute_service_up']= combined_updated['institute_service'].astype(str).str.extract(r'(\d+)')
```

```python
combined_updated['institute_service_up']=combined_updated['institute_service_up'].astype(float)
```

```python
combined_updated['institute_service_up']
```

```python
#create a formula that we will use to map the rest of the dataframe 
def years_comp(element):
    if element >= 11:
        return 'veteran'
    elif 10 > element >7:
        return 'established'
    elif 6 > element > 3:
        return 'experinced'
    elif pd.isnull(element):
        return np.nan
    else:
        return 'new'
```

```python
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(years_comp)
```

```python
combined_updated['service_cat']
```

```python
combined_updated['dissatisfied'].value_counts(dropna = False)
```

```python
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
```

```python
combined_updated['dissatisfied'].value_counts(dropna = False)
```

```python
pct_diss = combined_updated.pivot_table(index='service_cat',values = 'dissatisfied')
```
